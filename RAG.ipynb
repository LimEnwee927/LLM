```python
from langchain import LlamaCpp
!pip install llama-cpp-python
```

```python

llm = LlamaCpp(
    model_path = "Phi-3-mini-4k-instruct-fp16.gguf",
    n_gpu_layers = -1,
    max_tokens =  500,
    seed = 42, ## consistent output
    VERBOSE = False, ## see loading time, inference progress
    n_ctx=2048 ## model memory
    )
```

```python
!pip install langchain-community
from langchain.embeddings.huggingface import HuggingFaceBgeEmbeddings
```

```python
from sentence_transformers import SentenceTransformer
```

```python
## Embedding model to convert text to vector
embedding_model = SentenceTransformer('thenlper/gte-small')
embeddings = HuggingFaceBgeEmbeddings(model_name="thenlper/gte-small")
```

```python
text = """
Enwee is an Electrical and Electronic Engineering (EEE) student from Nanyang Technological University (NTU).
They have a strong interest in machine learning, particularly in the area of large language models (LLMs).
Enwee is passionate about building chatbots that are both efficient and accurate in providing responses.
In addition to their academic and professional interests, Enwee enjoys basketball.
"""

# Split into a list of sentences
texts = text.split('.')

# Clean up to remove empty spaces and new lines
texts = [t.strip(' \n') for t in texts]
```

```python
!pip install faiss-cpu
from langchain.vectorstores import FAISS
db = FAISS.from_texts(texts, embeddings)
```

```python
from langchain import PromptTemplate
from langchain.chains import RetrievalQA

template = """<|user|>
Relevant information :{context}
"Provide a concise answer to the following question based on the information above."
{question}


<|end|>
<|assistant|>"""


```

```python
prompt = PromptTemplate(template=template, input_variables=["context", "question"])
rag = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=db.as_retriever(), return_source_documents=True, chain_type_kwargs={"prompt": prompt}, verbose = True)
```

```python
rag.invoke('Who is Enwee')
```
